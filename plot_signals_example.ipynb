```python
import pandas as pd
import matplotlib.pyplot as plt

# Load waveforms
fast_signal_df = pd.read_csv("output/fast_signal.csv")
slow_signal_df = pd.read_csv("output/slow_signal.csv")

def save_signals_by_pmt(df):
    signals = {}
    grouped = df.groupby('KEY')
    
    for name, group in grouped:
        signals[name] = group['Signal'].values 
    return signals

fast_signals = save_signals_by_pmt(fast_signal_df)
slow_signals = save_signals_by_pmt(slow_signal_df)

plt.figure(figsize=(20, 10), dpi = 200)
plt.subplot(221)
plt.plot(fast_signals['time'] * 1e9, fast_signals['pmt_1'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 1')

plt.subplot(222)
plt.plot(fast_signals['time'] * 1e9, fast_signals['pmt_2'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 2')

plt.subplot(223)
plt.plot(fast_signals['time'] * 1e9, fast_signals['pmt_3'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 3')

plt.subplot(224)
plt.plot(fast_signals['time'] * 1e9, fast_signals['pmt_4'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 4')

plt.figure(figsize=(20, 10), dpi = 200)
plt.subplot(221)
plt.plot(slow_signals['time'] * 1e9, slow_signals['pmt_1'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 1')

plt.subplot(222)
plt.plot(slow_signals['time'] * 1e9, slow_signals['pmt_2'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 2')
plt.subplot(223)

plt.plot(slow_signals['time'] * 1e9, slow_signals['pmt_3'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 3')

plt.subplot(224)
plt.plot(slow_signals['time'] * 1e9, slow_signals['pmt_4'], 'k')
plt.grid()
plt.ylabel("Amplitude (V)")
plt.xlabel("Time (ns)")
plt.autoscale(True, 'x', True)
plt.title('PMT 4')
```
